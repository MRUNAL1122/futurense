```python
Function to print the Tic Tac Toe board
def print_board(board):
    for row in board:
        print(" | ".join(row))  # Print each row with '|' between elements
        print("-" * 9)  # Print a horizontal line between rows

# Function to check if there's a winner on the board
def check_winner(board):
    # Check rows
    for row in board:
        if len(set(row)) == 1 and row[0] != ' ':  # If all elements in a row are the same and not empty
            return row[0]  # Return the winner's symbol

    # Check columns
    for col in range(3):
        if len(set([board[row][col] for row in range(3)])) == 1 and board[0][col] != ' ':
            # If all elements in a column are the same and not empty
            return board[0][col]  # Return the winner's symbol

    # Check diagonals
    if len(set([board[i][i] for i in range(3)])) == 1 and board[0][0] != ' ':
        # If all elements in the main diagonal are the same and not empty
        return board[0][0]  # Return the winner's symbol
    if len(set([board[i][2 - i] for i in range(3)])) == 1 and board[0][2] != ' ':
        # If all elements in the other diagonal are the same and not empty
        return board[0][2]  # Return the winner's symbol

    return None  # Return None if no winner is found

# Function to check if the board is full
def is_board_full(board):
    return all(all(cell != ' ' for cell in row) for row in board)
    # Check if all cells in all rows are not empty (i.e., the board is full)

# Function to get a player's move (row and column)
def get_player_move():
    row = int(input("Enter row (0, 1, or 2): "))
    col = int(input("Enter column (0, 1, or 2): "))
    return row, col

# Main function for the Tic Tac Toe game
def tic_tac_toe():
    # Initialize an empty 3x3 board
    board = [[' ' for _ in range(3)] for _ in range(3)]
    # Start with player 'X'
    current_player = 'X'

    while True:
        # Print the current state of the board
        print_board(board)

        # Prompt the current player for their move
        print(f"Player {current_player}, it's your turn.")
        row, col = get_player_move()

        # Check if the chosen cell is empty
        if board[row][col] == ' ':
            # Place the current player's symbol in the chosen cell
            board[row][col] = current_player
            # Check if the current player has won
            winner = check_winner(board)

            if winner:
                print(f"Player {winner} wins!")
                break
            # Check if the board is full, resulting in a tie
            elif is_board_full(board):
                print("It's a tie!")
                break

            # Switch to the other player for the next turn
            current_player = 'O' if current_player == 'X' else 'X'
        else:
            # If the chosen cell is already taken, prompt the player to try again
            print("Cell already taken. Try again.")

# Run the Tic Tac Toe game when the script is executed
if _name_ == "_main_":
    tic_tac_toe()
```
